```python
# Импортируйте все необходимые библиотеки и модули которые вам необходимы
import pandas as pd
import numpy as np
```

```python
# Загрузите датасет и проведите его очистку и форматирование если это необходимо
df = pd.read_csv( './baseball_games.csv', delimiter = ',' )
df.head()
```

```python
data = df.drop(columns=['away_team_errors','field_type','venue','wind_direction','sky']).copy()
pd.set_option('display.max_columns', None)
data.head()
```

```python
# post season removed from the dataset
data.drop(data[data['season'] == 'post season'].index, inplace=True)
data.season.value_counts()
```

```python
data.loc[pd.isnull(data['attendance'])]
```

```python
# NaN values filled in with the mean value
mean_attendance = data['attendance'].mean()
data['attendance'].fillna((mean_attendance), inplace = True)
```

# 12. В какой игре было максимальное количество ранов?

_Note:_ max number of total **hits** in the game.

```python
data['total_hits'] = data[['away_team_hits', 'home_team_hits']].sum(axis = 1)
total_hits = data[data.total_hits == data.total_hits.max()]
total_hits
```

# 13. Выведите количество игр, которая сыграла каждая команда в данном сезоне?

```python
df1 = pd.DataFrame(data.away_team.value_counts(), columns=['away_team'])
df2 = pd.DataFrame(data.home_team.value_counts(), columns=['home_team'])
df3 = df1.join(df2)
df3['total_games'] = df3[['away_team', 'home_team']].sum(axis = 1)
df3.sort_index()
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

_Note:_ question 14 is below questions 15, 16

```python
# max number of home-team wins
home_win = data.groupby('home_team')[['home_team_win']].sum()
home_win.sort_values(by='home_team_win', ascending=False).head()
```

# 16. Какая команда выиграла наибольшее количество гостевых матчей в сезоне?

```python
# max number of away-team wins
away_win = data.groupby('away_team')[['home_team_loss']].sum()
away_win.sort_values(by='home_team_loss', ascending=False).head()
```

# 14. Какая команда выиграла наибольшое количество матчей в сезоне? <s>(Будьте внимательны с типом матча в сезоне - season).</s>

```python
# max wins within a season
total_wins = home_win.join(away_win)
total_wins['total_wins'] = total_wins[['home_team_win', 'home_team_loss']].sum(axis = 1)
total_wins.sort_values(by='total_wins', ascending=False).head()
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

```python
# max number of loss within a season
total_wins.set_index(['home_team_win'])
total_loss = total_wins.join(df3).sort_index()
total_loss['total'] = total_loss.total_games - total_loss.total_wins
total_loss[['total']].sort_values(by='total', ascending=False). head()
```
